```python
df_datasources = pd.read_csv("datasources_p.csv")
df_series = pd.read_csv("series_p.csv")
df_seriesdata = pd.read_csv("seriesdata_p.csv")
df_sourcesandseries = pd.read_csv("SourcesAndSeries.csv")
```

```python
df_series['seriesvariable_standardname'] = df_series['seriesvariable_standardname'].str.replace('(',' ', regex = False).str.replace(')', '', regex = False)
df_series[['name', 'type', 'period']] = df_series['seriesvariable_standardname'].str.split(' ', n = 2, expand = True)
df_series = df_series.drop('seriesvariable_standardname', axis=1)
```

```python
df_seriesdata = pd.merge(df_seriesdata, df_series[['name', 'type', 'period', 'unit_symbol', 'series_id', 'seriesvariable_longname']], on = 'series_id', how = 'left')
df_seriesdata
```
